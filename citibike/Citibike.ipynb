```python
flight_paths = []
for i in range(len(popular_trip_15)):
    flight_paths.append(
        go.Scattermapbox(
            lon = [popular_trip_15['start station longitude'][i], popular_trip_15['end station longitude'][i]],
            lat = [popular_trip_15['start station latitude'][i], popular_trip_15['end station latitude'][i]],
            mode = 'lines',
            line = dict(width = 5),
            opacity=float(popular_trip_15['tripduration'][i]) / float(popular_trip_15['tripduration'].max()) ,
            name = popular_trip_15['start station name'][i] + ' to ' + popular_trip_15['end station name'][i],
        )
    )
```

```python
layout = go.Layout(
    autosize=True,
    hovermode='closest',
    width=900,
    height=600,
    mapbox=go.layout.Mapbox(
        accesstoken='pk.eyJ1IjoiY2FvbWluaDA0MTIiLCJhIjoiY2p1bTRkMm0yMGs1ZzN5cXFnMWo2c2oxaCJ9.8q4cOzC4H-T3X35LgE_iVA',
        bearing=10,
        center=go.layout.mapbox.Center(
            lat=40.732219 ,
            lon=-73.981656
        ),
        style= "mapbox://styles/caominh0412/cjum4pbps2el31fmiolve2i4h",
        zoom=12
    ),
)
```

```python
station = [go.Scattermapbox(
                    lon = station_df['start station longitude'],
                    lat = station_df['start station latitude'],
                    text = station_df['start station name'],
                    hoverinfo = 'text',
                      mode = 'markers',
                    showlegend=False,
                   marker = go.scattermapbox.Marker(size=station_df['tripduration']/500))]
```

```python
fig = go.Figure(data=flight_paths+station ,layout=layout)
```
