```python
from sklearn.preprocessing import LabelEncoder
```

```python
df_copy['Platform'] = LabelEncoder().fit_transform(df_copy['Platform'])
```

```python
X=df_copy[['Critic_Score','Critic_Count','User_Score','User_Count','Platform']]
y=df_copy['Global_Sales']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR(gamma='scale', kernel='rbf', C=1)
svr.fit(X_train,y_train)
```

```python
predictions = svr.predict(X_test)
```

```python
test_df = X_test.copy()
test_df['sale'] = y_test
test_df['predict'] = predictions
```

```python
test_df.head(10)
```

```python
svr.score(X_test,y_test)
```

```python
svr.support_vectors_
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

xs = test_df.head(100)['Critic_Score']
ys = test_df.head(100)['User_Score']
zs = test_df.head(100)['sale']
zs2 = test_df.head(100)['predict']
ax.scatter(xs, ys, zs, s=20, alpha=0.3, edgecolors='w')
ax.scatter(xs, ys, zs2, s=20, alpha=0.6, edgecolors='r')


ax.set_xlabel('Critic_Score')
ax.set_ylabel('User_Score')
ax.set_zlabel('sale')

plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train,y_train)
```

```python
dtr_predictions = dtr.predict(X_test)
test_df2 = X_test.copy()
test_df2['sale'] = y_test
test_df2['predict'] = dtr_predictions
```
